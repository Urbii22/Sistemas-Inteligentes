El objetivo del juego es ocupar el mayor número posible de casillas, y que éste sea mayor que el número ocupado por el otro jugador. Tal y como comentamos en la clase de la semana anterior:

- La forma de aumentar el número de fichas que tiene un jugador es capturando las fichas del oponente. 


- Las fichas enemigas que queden en línea recta (vertical, horizontal o diagonal) entre la ficha del jugador activo y cualquiera de las otras fichas de éste, se *dan la vuelta* (es decir, cambian de color) siendo así *capturadas*.


- Sólo se permite poner fichas en las posiciones que no han sido ocupadas. 


- **IMPORTANTE**: En relación con el punto anterior, cada partida consta de hasta **60 turnos** (**64** casillas del tablero menos las **4** ocupadas inicialmente). 


- **NO** se pueden dejar huecos en el movimiento de una ficha entre la del oponente y la casilla **a la que vas**.


- Sólo se puede poner una ficha si se capturan fichas del oponente. De esta forma, sólo se pueden poner fichas en las casillas adyacentes a otras fichas. 


- Si no se puede poner ficha, se pasa el turno.


- La partida llega a su fin cuando:


        Se ocupan las 64 casillas del tablero.

        Ninguno de los dos jugadores puede mover.

En esta práctica, vamos a:

1) Definir una función que nos va a determinar el número de fichas que se dan la vuelta respecto de una posición $(x,y)$ determinada por un jugador.

2) Entender y establecer una primera **función de coste** o **heurística** a aplicar para que el juego pueda ser optimizado.

## Generación de una función que determine el número de fichas que se dan la vuelta

Para ello, hemos de hacer una *búsqueda* en las 8 posibles direcciones que se pueden definir a partir de una determinada casilla. Es decir:

```python
# ==============================================================================================================================
#This method checks the game rules.
# 
# I Understand that "isAvailable" returns the number of available moves that can be made. If it is "0", then NO MOVEMENT
# can be made.

# In fact, it returns the variable "count"
#
# "count" variable is accumulated throughout the code. In a part of this code ("isAvailable"), it searches through all the 
# diagonal directions starting from a single board position.
#
# It is curious to find that PLAYMODE never changes once it is declared in the argument inside the parenthesis in the "def".
# ==============================================================================================================================

def isAvailable(board, row, col, playerID, PLAYMODE=True):
    global changed
    global player
    global debug
    global victory
    global whiteTiles
    global blackTiles

    if PLAYMODE:
        board[row][col] = player
    count = 0
    __column = board[row]                        # This sentence gives the values in row "row", in the list of lists (matrix)
    __row = [board[i][col] for i in range(0, 8)] # This sentence gives the values in column "col", in the 
                                                 # list of lists (matrix)
    
    if playerID in __column[:col]:      # Here we identify the first "col" number of elements in "__column", i. e., it obtains 
                                        # the first "col" elemtns in the row 
        changes = []
        searchCompleted = False             # By default,"searchCompleted = False", meaning that a search has to be made.
        for i in range(col - 1, -1, -1):    # Here, "i" goes from "col-1" up to "0", both of them included.
            if searchCompleted:
                continue
            piece = __column[i]
            if piece == 0:
                changes = []
                searchCompleted = True
            elif piece == playerID:
                searchCompleted = True
            else:
                changes.append(i)  # An "append" is made if the position "i" in the row is such that it is not empty ("0") AND
                                   # it does not contain a chip/tile with the same ID as "playerID". I understand this means
                                   # that it identifies the positions in the row that belong to THE OTHER PLAYER.
                        
                                   # Here, (in "else") we do not change the "searchCompleted" variable value, meaning that it
                                   # will continue SEARCHING.
                        
        if searchCompleted:        # If it is TRUE, count the number of changes that would IMPLY a movement. THis would be, on
                                   # some occasions, the best strategy.
            count += len(changes)
            if PLAYMODE:
                for i in changes:
                    board[row][i] = player


    if playerID in __column[col:]:
        changes = []
        searchCompleted = False

        for i in range(col + 1, 8, 1): # Here, "i" goes from "col+1" up to "7", both of them included.
            if searchCompleted:
                continue
            piece = __column[i]
            if piece == 0:
                changes = []
                searchCompleted = True
            elif piece == playerID:
                searchCompleted = True
            else:
                changes.append(i)  # An "append" is made if the position "i" in the row is such that it is not empty ("0") AND
                                   # it does not contain a chip/tile with the same ID as "playerID". I understand this means
                                   # that it identifies the positions in the row that belong to THE OTHER PLAYER.
                        
                                   # Here, (in "else") we do not change the "searchCompleted" variable value, meaning that it
                                   # will continue SEARCHING.
                                
        if searchCompleted:        # If it is TRUE, count the number of changes that would IMPLY a movement. THis would be, on
                                   # some occasions, the best strategy.
            count += len(changes)
            if PLAYMODE:
                for i in changes:
                    board[row][i] = player # All the positions in "changes" are changed to "player", taking into account that 
                                           # they belonged to the other player.



    if playerID in __row[:row]:    # Here we identify the first "row" number of elements in "__row", i. e., it obtains the first
                                   # "row" elements in a specific column of the matrix
        changes = []
        searchCompleted = False

        for i in range(row - 1, -1, -1): # Here, "i" goes from "row-1" up to "0", both of them included.
            if searchCompleted:
                continue
            piece = __row[i]
            if piece == 0:
                changes = []
                searchCompleted = True
            elif piece == playerID:
                searchCompleted = True
            else:
                changes.append(i)
        if searchCompleted:
            count += len(changes)
            if PLAYMODE:
                for i in changes:
                    board[i][col] = player



    if playerID in __row[row:]:
        changes = []
        searchCompleted = False

        for i in range(row + 1, 8, 1): # Here, "i" goes from "row+1" up to "7", both of them included.
            if searchCompleted:
                continue
            piece = __row[i]
            if piece == 0:
                changes = []
                searchCompleted = True
            elif piece == playerID:
                searchCompleted = True
            else:
                changes.append(i)
        if searchCompleted:
            count += len(changes)
            if PLAYMODE:
                for i in changes:
                    board[i][col] = player


#   [row - i][col - i]

    i = 1    
    ulDiagonal = []
    while row - i >= 0 and col - i >= 0:
        ulDiagonal.append(board[row - i][col - i])
        i += 1
    if playerID in ulDiagonal:
        changes = []
        searchCompleted = False
        for i in range(0, len(ulDiagonal)):
            piece = ulDiagonal[i]
            if searchCompleted:
                continue
            if piece == 0:
                changes = []
                searchCompleted = True
            elif piece == playerID:
                searchCompleted = True
            else:
                changes.append((row - (i + 1), col - (i + 1)))  # Here we will also include in "changes", those cells 
                                                                # in the Upper Left diagonal of the board
        if searchCompleted:
            count += len(changes)
            if PLAYMODE:
                for i, j in changes:
                    board[i][j] = player


#   [row + i][col - i]

    i = 1
    urDiagonal = []
    while row + i < 8 and col - i >= 0:
        urDiagonal.append(board[row + i][col - i])
        i += 1
    if playerID in urDiagonal:
        changes = []
        searchCompleted = False
        for i in range(0, len(urDiagonal)):
            piece = urDiagonal[i]
            if searchCompleted:
                continue
            if piece == 0:
                changes = []
                searchCompleted = True
            elif piece == playerID:
                searchCompleted = True
            else:
                changes.append((row + (i + 1), col - (i + 1)))   # Here we will also include in "changes", those cells 
                                                                 # in the Lower Left diagonal of the board
        if searchCompleted:
            count += len(changes)
            if PLAYMODE:
                for i, j in changes:
                    board[i][j] = player

                    
#   [row - i][col + i]

    i = 1
    llDiagonal = []
    while row - i >= 0 and col + i < 8:
        llDiagonal.append(board[row - i][col + i])
        i += 1
    if playerID in llDiagonal:
        changes = []
        searchCompleted = False

        for i in range(0, len(llDiagonal)):
            piece = llDiagonal[i]
            if searchCompleted:
                continue
            if piece == 0:
                changes = []
                searchCompleted = True
            elif piece == playerID:
                searchCompleted = True
            else:
                changes.append((row - (i + 1), col + (i + 1)))   # Here we will also include in "changes", those cells 
                                                                 # in the Upper Right diagonal of the board
        if searchCompleted:
            count += len(changes)
            if PLAYMODE:
                for i, j in changes:
                    board[i][j] = player

                    
#   [row + i][col + i]


    i = 1
    lrDiagonal = []
    while row + i < 8 and col + i < 8:
        lrDiagonal.append(board[row + i][col + i])
        i += 1
    if playerID in lrDiagonal:
        changes = []
        searchCompleted = False

        for i in range(0, len(lrDiagonal)):
            piece = lrDiagonal[i]
            if searchCompleted:
                continue
            if piece == 0:
                changes = []
                searchCompleted = True
            elif piece == playerID:
                searchCompleted = True
            else:
                changes.append((row + (i + 1), col + (i + 1)))   # Here we will also include in "changes", those cells 
                                                                 # in the Lower Right diagonal of the board

        if searchCompleted:
            count += len(changes)
            if PLAYMODE:
                for i, j in changes:
                    board[i][j] = player

    if count == 0 and PLAYMODE:
        board[row][col] = 0

    return count
```

# Definición de una primera/potencial *función de coste* o *heurística*

La primera función de coste que os propongo va a ser una función que constará de dos términos, tal que:

$U$ = $\delta p$ + $M$

donde:

- $U$: Función de *utilidad*

- $\delta p$: Porcentaje de ocupación

- $M$: Movilidad

El porcentaje de ocupación se puede definir de dos formas:

1) $\delta p = \left(\frac{N_{w}}{N_{w}+N_{b}}\right) \times 100$

2) $\delta p = - \left(\frac{N_{b}}{N_{w}+N_{b}}\right) \times 100$

donde:

- $N_{b}$: Número de casillas ocupadas por ficha con la **cara negra** visible

- $N_{w}$: Número de casillas ocupadas por ficha con **cara blanca** visible

Por tanto, esta función tendrá un valor positivo, o negativo, en el intervalo $[0, 100]$

En cuanto a la función $M$, se define como:

$M = \left(\frac{1_{M(2)}}{1_{M(2)} + 1_{M(1)}}\right) \times 100$

En esta ecuación:

- $1_{M(1)}$: Función que devuelve $0$ o $1$, dependiendo de si el movimiento de la casilla ocupada por $1$ (casilla con cara **negra**) se puede llevar a cabo.

- $1_{M(2)}$: Función que devuelve $0$ o $1$, dependiendo de si el movimiento de la casilla ocupada por $2$ (casilla con cara **blanca**) se puede llevar a cabo.

Con esta definición, se ha de evitar que $M$ pueda llegar a tener valor de indeterminación numérica. En este caso, podría ocurrir tener: $\frac{0}{0}$. 

Por tanto:

- Si $1_{M(1)}=1_{M(2)}=0 $, consideraremos que $M = 0$
